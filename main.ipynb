```python

from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_google_genai import GoogleGenerativeAIEmbeddings
import faiss
from langchain_community.vectorstores import FAISS

embeddings_model = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview", task_type="RETRIEVAL_DOCUMENT")
index = faiss.IndexFlatL2(len(embeddings_model.embed_query("hello world")))

vector_store=FAISS(docstore=InMemoryDocstore(), embedding_function=embeddings_model, index=index, index_to_docstore_id={})
vector_store.add_documents(split_docs)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash" , temperature=0)
```

```python

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser


template=""" Use the below retrieved context to answer the question and answer , if you dont know answer just say you are not aware of the answer. Provide a concise answer.

Context: {context}
Question:{q}
Answer: 
"""

prompt=ChatPromptTemplate.from_template(template)

rag_chain=(
    {'context' : lambda x:results[0].page_content,  'q' : RunnablePassthrough()}
    |prompt
    |llm
    |StrOutputParser()
    
)
q="When did I  passed Bachelor of Technology from Coochbehar Governement Engineering college?"
r=rag_chain.invoke(q)
print(r)
```

```python
from langsmith import Client

client= Client()
examples=[{
    "inputs" : {"question" : "What are your expertise?"},
    "outputs" : {"answer" : "GenAI, RAG , LLM , AWS "}
},
{
    "inputs" : {"question" : "When did you complete your graduation?"},
    "outputs" : {"answer" : "2022"}
},
{
    "inputs" : {"question" : "When did you start pursuing college in Coochbehar Government Engineering College?"},
    "outputs" :{"answer" : " 2018"}
}]


dataset_name="checker"
dataset= client.create_dataset(dataset_name=dataset_name)
client.create_example(
    dataset_id=dataset.id,
    examples=examples
)
```
